```python
# Escrever um programa que recebe as notas dos 04 biomestre e imprimi a média com o resultado de aprovado, recuperação ou reprovado
# Regras: Utilizar while,
# nota igual ou maior que 7 : APROVADO
# nota igual ou maior que 5 recperação
# Nota menor que 5 reprovado.

nota = 0
bimestre = 1
resultado = 0

while bimestre <= 4:
  nota = float (input(f"digite a sua nota do {bimestre}° bimestre: "))
  bimestre += 1
  resultado = resultado + nota

if (resultado /4) >= 7:
  print (f" Sua média foi {resultado/4}, você está aprovado")
elif (resultado /4) >= 5:
  print (f"Sua média foi {resultado/4}, você está de recuperação")
else :
  print (f"Sua média foi {resultado/4}, você está reprovado")
```

Projeto caixa eletrônica

Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

O sistema deverá funcionar atravéz de um menu interativo utilizando "while"

O programa deve inicar com um saldo inicial de R$1000,00 e permitir que o usuário escolham diferentes operações até decidir sair do sistema


```python
saldo = 1000
opcao = 0

while opcao != 4:
    print("    CAIXA ELETRÔNICO   ")
    print("opção 1: Consultar saldo")
    print("opção 2: Depositar")
    print("opção 3: Sacar")
    print("opção 4: Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        print(f"Seu saldo atual é: R${saldo}")

    elif opcao == 2:
        deposito = float(input("Digite o valor do depósito: R$"))

        if deposito > 0:
            saldo += deposito
            print("Depósito realizado com sucesso!")
        else:
            print("Valor inválido.")

    elif opcao == 3:
        saque = float(input("Digite o valor do saque: "))

        if saque <= saldo:
            saldo -= saque
            print("Saque realizado com sucesso!")
        else:
            print("Saldo insuficiente.")

    elif opcao == 4:
        print("Saindo do sistema")

    else:
        print("Opção inválida. Tente novamente")

print("Estarei a sua disposição!")
```
