```python
# Questão 4
energia = float(input("Digite a quantidade de energia que tem o seu personagem: "))

if energia == 0:
  print (f"De acordo com a energia, que seria {energia}, seu personagem esta morto")


elif energia >= 1 and energia <= 30:
  print(f"De acordo com a energia, que seria {energia}, seu personagem esta critico")

elif energia >= 31 and energia <= 70:
  print (f"De acordo com a energia, que seria {energia}, seu personagem esta normal")

else:
  print(f"De acordo com a energia, que seria {energia}, seu personagem esta saudável")

```
