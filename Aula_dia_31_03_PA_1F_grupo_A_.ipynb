```python
# Estrutura de decisões

# Escrever um programa que recebe o nome e a idade do usuário e verifica se o usuario é criança, adolescente e adulto

# >= 12 anos
# =< 17 anos é adolescentes
# Adulto

# Entrada

nome = str (input ("Digite seu nome: "))
idade = int (input ("Digite sua idade: "))

# Processamento / Saida

if idade <= 12:
  print (f"{nome} voce tem {idade} anos de idade, voce é criança.")

elif idade <= 18:
  print (f"{nome} você tem {idade} de idade, voce é adolescente]")

else:
  print (f"{nome} voce tem {idade} anos de idade, voce é adulto.")

```

```python
#Escrever um programa que classifica os tipos de um triângulo e valida se é um triângulo

# Validar se os lados digitados pode formar um triângulo
# lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1

# Tipos de triângulo
# Equilatero - Todos os lados iguais - ==
# Escaleno - Todos os lados diferentes - !=
# Isósceles = dois lados iguais

lado1 = float (input("Digite o valor do primeiro lado: "))
lado2 = float (input("Digite o valor do segundo lado: "))
lado3 = float (input("Digite o valor do terceiro lado: "))

# Processamento / Saida

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1 :
    print ("É um triângulo valido")

    if lado1 == lado2 == lado3 :
        print ("É um triângulo equilátero")

    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3 :
        print ("É um triângulo isósceles")

    else:
        print ("É um triângulo escaleno")

else:
    print ("Não é um triangulo valido")
```
