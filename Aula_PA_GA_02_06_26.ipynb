```python
# Fazer um jogo de adivinhação

# O usuário ira digitar um numero de 0 a 10 para adivinhar o numero secreto
# O usuário terá 03 tentativas para acertar o numero
# O programa deverá imprimir uma mensagem de quente, morno ou frio

# Regras

# Se a difença do palpite para o número secreto for 3- frio, 2- morno e 1- quente

import random


numero_secreto = random.randint(0, 10)
tentativas = 3

print("--- JOGO DA ADIVINHAÇÃO ---")
print("Tente adivinhar o número secreto entre 0 e 10.")
print(f"Você tem {tentativas} tentativas. Boa sorte!")

for tentativa in range(1, tentativas + 1):

    try:
        palpite = int(input(f"Tentativa {tentativa}/{tentativas} - Digite seu palpite: "))
    except:
        print("Por favor, digite um número válido.")


    if palpite > numero_secreto:
        diferenca = palpite - numero_secreto
    else:
        diferenca = numero_secreto - palpite


    if diferenca == 0:
        print(f"Parabéns! Você acertou! O número secreto era {numero_secreto}.")

    elif diferenca == 1:
        print(" Quente!")
    elif diferenca == 2:
        print(" Morno!")
    elif diferenca >= 3:
        print(" Frio!")

else:
    print(f" Fim de jogo! Suas tentativas acabaram. O número secreto era {numero_secreto}.")
```
