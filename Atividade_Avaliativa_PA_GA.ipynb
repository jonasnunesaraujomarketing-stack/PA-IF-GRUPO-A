```python
# Questão 2

numero = int(input("Digite um número inteiro: "))

soma = 0

for i in range(1, numero + 1):
    soma += i

print(f"Resultado:{soma}")
```

```python
# Questão 3

numero = int(input("Digite um número: "))

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

```python
# Questão 4

numero = -1
while numero < 0:
    numero = int(input("Digite um número inteiro positivo: "))
    if numero < 0:
        print("Número inválido! Por favor, digite um número maior ou igual a zero.")

fatorial = 1

for i in range(2, numero + 1):
    fatorial *= i

print(f"Fatorial = {fatorial}")
```

```python
#Questão 5

saque = int(input("Digite o valor do saque: "))


for valor_nota in (100, 50, 20, 10, 5, 2, 1):
    quantidade_notas = 0


    while saque >= valor_nota:
        saque -= valor_nota
        quantidade_notas += 1


    if quantidade_notas > 0:
        print(f"{quantidade_notas} notas de R${valor_nota}")
```
