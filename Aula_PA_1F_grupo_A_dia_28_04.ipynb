```python
# Escrever um programa que o usario digita uma palavra, e o programa verifica se a palavra é palindroma e imprima a palavra

#Entrada
palavra = str (input("Digite uma palavra: "))
invertida = ""

# Processamento
for i in (palavra):

 invertida = i + invertida

if palavra == invertida:
  print(f"A palavra {palavra} é palindroma")
else:
  print(f"A palavra {palavra} não é palindroma")

```
