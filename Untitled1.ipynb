```python
#Exercicio 1 Escrever um programa que o usuário digita um número e o programa imprimi até o número digitado.
# Entrada

numero = int(input("Digite um número: "))

for i in range(0, numero + 1):
    if i % 1 == 0:
        print(i)
```

```python
#Exercicio 2
#Escrever um programa que o usuário digita um número e o programa imprimi o número digitado até o zero (contagem regressiva).
#Entrada
numero = int(input("Digite um número: "))

for i in range (numero, -1,-1):
    print(i)
```

```python
# 3 - Escrever um programa que o usuário digita um número e o programa imprimi os números pares.
n1 = int(input("Digite um número: "))

for i in range(0, n1 + 2):
    if i % 2 == 0:
        print(i)
```

```python
# 4 - Escrever um programa que o usuário digita um número e o programa imprimi os números impares.
# Entrada
n1 = int(input("Digite um número: "))

for i in range(0, n1 + 2):
    if i % 2 != 0:
        print(i)
```

```python
# 5 - Escrever um programa que o usuário digita um número e programa soma os números até o número digitado.
# Entrada

n1 = int(input("Digite um número: "))
soma = 0

for i in range(n1 + 1):
    soma += i
print(f"A soma até o numero digitado é: {soma}")
```
