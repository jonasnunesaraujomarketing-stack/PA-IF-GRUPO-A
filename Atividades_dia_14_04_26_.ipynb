```python
#Exercicio 1
# Par ou impar
# Entrada
n1 = float(input("digite um numero"))
#processamento/Saida
if n1 % 2 == 0:
  print (f"O numero {n1} é par")
else:
  print (f"O numero {n1} é impar")
```

```python
#Exercicio 2
n1 = float(input("digite um numero"))
n2 = float(input("digite um numero"))
#processamento
if n1 > n2:
  print (f"O numero {n1} é maior")
elif n1 < n2:
  print (f"O numero {n2} é maior")
else:
  print (f"Os numeros são iguais")
```

```python
#Exercicio 7
#Entrada
n1 = float(input("digite um numero: "))
n2 = float(input("digite outro numero: "))
processamento = str(input("digite um processamento que você deseja (+ - / *): " ))

#processamento

if processamento == '+' :
 print(f"A soma dos numeros é {n1+n2}")

elif processamento == "-":
  print(f"A subtração dos numeros é {n1-n2}")

elif processamento == "*":
   print(f"A multiplicação dos numeros é {n1*n2}")

elif processamento == "/":
    print(f"A divisão dos numeros é {n1/n2}")
```
