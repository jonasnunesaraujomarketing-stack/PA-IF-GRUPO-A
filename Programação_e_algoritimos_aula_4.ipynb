Calcular a área do retangulo

```python
# Entrada
comprimento = float(input("Digite o comprimento do retangulo:"))
Altura = float (input("Digite a altura do retângulo"))


# Processamento

area = comprimento * Altura


# Saida

print (f"O total da Area do retângulo é:{area}")
```

Fazer um programa que calcula o dobro de um numero

Entrada: um numero

Processamento: Calcular o dobro

Saida : mostrar o resultado

```python
# Entrada

numero = float(input("Coloque o numero que voce quer em dobro:"))

# Processamento

Total = numero * 2

# saida

print (f"O resultado do {numero} que voce quer em dobro é : {Total}" )
```

Escrever um programa que calcula a media das notas dos alunos
Entrada: nome, idade e as 4 notas do bimestre
Processamento: Calcular a media
Saida: mostrar o nome do aluno e a média obtida

```python
# entrada
nome = str(input('Escreva seu nome'))

idade = int(input('Qual a sua idade?'))

Responda = str(input("Coloque a média que voce obtive ok?" ))

primeiro = float(input("Média do primeiro bimestre:"))

segundo = float(input('Média do segundo bimestre'))

terceiro = float(input('Média do terceiro bimestre'))

quarto = float(input('Média do quarto bimestre'))

# processamento

Total = (primeiro + segundo + terceiro + quarto) / 4

# saida

print (f'Olá {nome} a média de todos os seus bimestres é : {Total}')
```
